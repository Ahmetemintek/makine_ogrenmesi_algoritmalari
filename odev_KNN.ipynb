```python
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
y=df3["winsorize_life_expectancy"]
X= df3[request_col]
X_train, X_test, y_train, y_test= train_test_split(X,y, random_state=111, test_size=0.2)

for i in range(5,20,2):
    model_capraz= KNeighborsRegressor(n_neighbors=i)
    model_capraz.fit(X_train, y_train)
    print("KomÅu sayÄ±sÄ± {} iÃ§in model skoru: {}".format(i, model_capraz.score(X_test, y_test)))
```

```python
cvscore= cross_validate(estimator=model_capraz, X=X_test, y=y_test, cv=10, return_train_score=True) #model yeniden oluÅturulup yapÄ±lacak
print("EÄitim skorlarÄ±: \n {}".format(cvscore["train_score"]))    #x_test ve y_tes ile
print("Test skorlarÄ±: \n {}".format(cvscore["test_score"]))
print("*"*50)
print("EÄitim skorlarÄ± ortalamasÄ±: {}".format(np.mean(cvscore["train_score"])))
print("Test skorlarÄ± ortalamasÄ±: {}".format(np.mean(cvscore["test_score"])))
```

Cross_val_score ile Ã¶lÃ§Ã¼m:

```python
for i in range(3,15):
    cv_score= cross_val_score(model_capraz, X_test, y_test, cv=i)
    print("CV:{} iÃ§in Ã§apraz doÄrulama skoru: {}".format(i, np.mean(cv_score)))
```
